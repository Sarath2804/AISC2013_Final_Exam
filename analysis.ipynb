```python
# Select the 10 important features
important_features = ['Life_expectancy', 'Adult_Mortality', 'infant_deaths', 'percentage_expenditure', 
                      'Hepatitis_B', 'Measles', 'BMI', 'Polio', 'GDP', 'Schooling']

# Filter the DataFrame to include only the important features
selected_df = df[important_features]

# Calculate the correlation matrix
correlation_matrix = selected_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Important Features')
plt.show()
```
